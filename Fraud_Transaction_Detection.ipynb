## **Importing the dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## **Data Processing**

```python
# reading csv file to pandas dataframe
credit_card_df = pd.read_csv('/content/creditcard.csv')
```

```python
# print first 5 rows
credit_card_df.head()
```

```python
# print last 5 rows
credit_card_df.tail()
```

```python
# get the information about the dataset
credit_card_df.info()
```

## **Finding Missing Values**

```python
credit_card_df.isnull().sum()
```

## **Distribution of normal transactions and fraudulent transactions**

```python
# 0 - Normal Transaction
# 1 - Fraudulent Transaction

credit_card_df['Class'].value_counts()
```

## **Data Analysis**

```python
normal_trans = credit_card_df[credit_card_df.Class == 0]
fraud_trans = credit_card_df[credit_card_df.Class == 1]
```

```python
# count of rows and cols
print(normal_trans.shape)
print(fraud_trans.shape)
```

## **Statistical Analysis**

```python
normal_trans.Amount.describe()
```

```python
fraud_trans.Amount.describe()
```

## **Finding the Difference**

```python
credit_card_df.groupby('Class').mean()
```

## **Random Sampling**

```python
normal_sample = normal_trans.sample(n=492)
```

## **Concatenate Normal and Fraud DataFrames**

```python
new_credit_card_df = pd.concat([normal_sample, fraud_trans], axis=0)
```

```python
new_credit_card_df.head()
```

```python
new_credit_card_df.tail()
```

## **Uniform Data Distribution**

```python
new_credit_card_df['Class'].value_counts()
```

```python
new_credit_card_df.groupby('Class').mean()
```

## **Separating the data and label**

```python
X = new_credit_card_df.drop(columns='Class', axis=1)
Y = new_credit_card_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

## **Splitting the data into Train data and Test Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
```

```python
print(X_test)
```

## **Training the Machine Learning Model**

#### Logistic Regression

```python
model = LogisticRegression(solver='liblinear', max_iter=1000)
model.fit(X_train, Y_train)
```

## **Model Evaluation**

#### Accuracy score on training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score on the training data: ', training_data_accuracy)
```

#### Accuracy score on test data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on the test data: ', test_data_accuracy)
```
